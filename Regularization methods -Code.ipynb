We removed only 17 items by just looking at the histograms. Let's see the count by using IQR method.<br>

IQR Outlier : We use cutoff at 1.5 x IQR (appx 95% curve)<br>

```python
q25,q75 = round(np.percentile(data_strip.iloc[:,0],25),2),round(np.percentile(data_strip.iloc[:,0],75),2)
print ('IQR = ',q75,'-',q25,'=',round(q75-q25,2))
iqr = q75-q25
cut_off = round(iqr*1.5,2)
cut_off_strong = round(iqr*3,2)
lower,upper = round(q25-cut_off,2),round(q75+cut_off,2)
lower_strong,upper_strong = round(q25-cut_off_strong,2),round(q75+cut_off_strong,2)
print('Lower:',lower,'Upper:',upper,'\nStrong lower i.e. 3*IQR:',lower_strong,'\n Strong Upper:',upper_strong)
outliers = [x for x in data.iloc[:,0] if x<lower or x>upper]
outliers_strong = [x for x in data.iloc[:,0] if x<lower_strong or x>upper_strong]
print ('Total Outliers', len(outliers),'\nStrong outliers', len(outliers_strong)) 
```

```python
#Lets write function to generate dummy
#Idea is to input the dataframe and a list of column locations having categorical variables 
def gen_dummy(x,list_var):
    for i in list_var:
        df = pd.get_dummies(x.iloc[:,i],prefix = x.columns[i],drop_first=True)          #create dummy of that particular variable
        x = pd.concat([x,df],axis = 1)  #concat the original df (i.e.'x') with this new df
    df_new=x
    for i in list_var:
        print ('Removing',x.columns[i])
        df_new = df_new.drop([x.columns[i]],axis=1)#drop the original column
    return df_new
        
```

### Defining the divide function to split the data. 


```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Divide the data into training and testing
def divide(data_tosplit,target_column_number,proportion):
    data_targets = data_tosplit.iloc[:,target_column_number]
    data_train = data_tosplit.drop(data_tosplit.columns[target_column_number],axis=1)
    x_train,x_test,y_train,y_test=train_test_split(data_train,data_targets,test_size=proportion,random_state=123)
    return x_train,x_test,y_train,y_test
```

```python
x_train,x_test,y_train,y_test = divide(data_x,0,0.4)
```

```python
x_train.head(2)
```

### Scaling

```python
print(x_train.iloc[:,:2].head(2))
print(x_test.iloc[:,:2].head(2))
```

We don't use the complete dataset because we don't want to leak the test data. For example, if train data has highest value
of 2000 and test has a highest value of 3000. If we scale using complete data, the model will know beforehand that there
is a value of 3000.

```python

x_train_scaled =x_train.copy()
x_test_scaled = x_test.copy()
for i in x_train.columns[:2]:
    print ('Scaling ',i)
    scaler = preprocessing.StandardScaler()
    x_train_scaled[i]=scaler.fit_transform(x_train_scaled[i].values.reshape(-1,1))
    x_test_scaled[i]=scaler.fit_transform(x_test_scaled[i].values.reshape(-1,1))
```

```python
print(x_train.iloc[:,:2].head(2))
print(x_test.iloc[:,:2].head(2))
print(x_train_scaled.iloc[:,:2].head(2))
print(x_test_scaled.iloc[:,:2].head(2))
```

### Defining a result function for all the methods

```python
#Result function
def result(x_train,model,y_test,predicted):
    t = pd.DataFrame(list(zip(x_train.columns,model.coef_)))
    print('Intercept is: %.2f'%model.intercept_)
    print('Coefficients are: \n',round(t,2),'\n')
    print("Mean squared error: %.2f" % round(metrics.mean_squared_error(y_test,predicted),2),'\n')
    print('Variance score: %.2f' % metrics.r2_score(y_test, predicted),'\n')
    predictors = x_train.columns
    coef= pd.Series(model.coef_,predictors).sort_values()
    coef.plot(kind='bar')
    plt.show()
    return (metrics.r2_score(y_test, predicted))
```

### Linear Regression

```python
#Linear regression
from sklearn import linear_model
from sklearn import ensemble
```

```python
def linear_reg(x_train,x_test,y_train,y_test):
    print ('RESULTS FOR LINEAR REGRESSION \n')
    model = linear_model.LinearRegression()
    model.fit(x_train,y_train)
    predicted = model.predict(x_test)
    result(x_train,model,y_test,predicted)
```

```python
linear_reg(x_train,x_test,y_train,y_test)
```

<a id="ridge"></a>
### Ridge Regression

```python
def ridge_reg(x_train,x_test,y_train,y_test,alpha):
    print ('RESULTS FOR Ridge REGRESSION \n')
    model = linear_model.Ridge(alpha = alpha)
    model.fit(x_train,y_train)
    predicted = model.predict(x_test) 
    result(x_train,model,y_test,predicted)
```

Using alpha from 0 to 10 in increments of 0.5

```python
for i in np.linspace(0,10,20):
    print (i)
    ridge_reg(x_train,x_test,y_train,y_test,i)
```

### Lasso

For Lasso, scaling is important since it directly puts the constraint on the size of the coefficients, which in turn depends on the magnitude of each value.

```python
print('Training \n',x_train_scaled.head(1),'\n\n\n','Testing \n',x_test_scaled.head(1),'\n')
```

```python
def lasso_reg(x_train,x_test,y_train,y_test,alpha):
    print ('RESULTS FOR Lasso REGRESSION \n')
    model = linear_model.Lasso(alpha = alpha)
    model.fit(x_train,y_train)
    predicted = model.predict(x_test)
    result(x_train,model,y_test,predicted)
```

```python
lasso_reg(x_train_scaled,x_test_scaled,y_train,y_test,0.001)
```

### Trying Grid Search CV for hyper parameter

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

Even though linear regression has no hyperparameters, the application of cross validation is to check for overfitting.

```python
lin_reg_1 = linear_model.LinearRegression()
MSEs = cross_val_score(lin_reg_1,x_train,y_train,cv=5)
mean_MSE = np.mean(MSEs)
print(MSEs)
print ('Mean MSE',mean_MSE)
print('Std. dev: %.2f'%(MSEs.std()*100),'%')
```

```python
ridge = linear_model.Ridge()
li =np.linspace(0,10,200)
parameters = {'alpha':li}
ridge_regressor = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(x_train,y_train)
```

```python
lasso = linear_model.Lasso()
li =np.linspace(0,1,200)
parameters = {'alpha':li}
lasso_regressor = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
lasso_regressor.fit(x_train_scaled,y_train)
```

```python
print("Hyper parameters and scores for Ridge regression \n")
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
print("\n Hyper parameters and scores for Lasso regression \n")
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```
